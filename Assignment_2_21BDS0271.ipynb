**Task 3.Perform the Below Visualizations.**

● Univariate Analysis

```python
sns.boxplot(df['Price'])
```

```python
sns.displot(df['Built Year'])
```

```python
plt.title('Number of schools nearby')
plt.pie(df['Number of schools nearby'].value_counts(),[0.05,0.05,0.05],labels=df['Number of schools nearby'].unique(),autopct='%1.1f%%',colors=['beige','pink','yellow'])
plt.show()
```

● Bi-Variate Analysis

```python
sns.scatterplot(x=df['Area of the house(excluding basement)'],y=df['Price'])
```

```python
sns.lineplot(x=df['living area'],y=df['Price'])
```

```python
sns.scatterplot(x=df['Area of the basement'],y=df['Price'])
```

● Multivariate Analysis

```python
rcParams['figure.figsize']=30,30
sns.heatmap(df.corr())
```
